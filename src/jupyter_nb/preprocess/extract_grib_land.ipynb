```python
import platform
import xarray as xr
import pandas as pd
import numpy as np
import os
import glob
```

```python
netcdf_dir = 'D:\\_datasets\\weather-dataset\\ERA5-Land-data\\land'
```

```python
netcdf_pattern = os.path.join(netcdf_dir, "*.grib")
netcdf_files = glob.glob(netcdf_pattern)
print(f"Found {len(netcdf_files)} GRIB files")
```

```python
all_files = []
for netcdf_file in netcdf_files:
    all_files.append(netcdf_file)
print(all_files[0])
print(all_files[1])
```

```python
backend_kwargs = {'indexpath': ''}
ds = xr.open_dataset(all_files[0], engine='cfgrib', backend_kwargs=backend_kwargs)
```

```python
print("\n=== Grib File Information ===")
print(f"Dimensions: {dict(ds.sizes)}")
print(f"Coordinates: {list(ds.coords.keys())}")
```

```python
vars_list = list(ds.data_vars.keys())
print(f"Dataset variables: {vars_list}")
```

```python
# Show details for up to first 5 variables
for var_name in vars_list:
    da = ds[var_name]
    print(f"  Variable '{var_name}': shape={da.shape} dims={da.dims}")
```

```python
# Show details variables
for var_name in vars_list:
    da = ds[var_name]
    print(f"  Variable '{var_name}': shape={da.shape} dims={da.dims}")
    for attr in da.attrs:
        print(f"    {attr}: {da.attrs[attr]}")
```
